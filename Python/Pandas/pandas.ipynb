## Filtering Data

```python
# find all names containing Mega
df.loc[df['Name'].str.contains('Mega')]
```

## More Advanced Filtering with Regular Expression

```python
df.loc[df['Type 1'].str.contains('Fire|grass', regex=True)]
```

## Aggregate Statistics (Groupby)

```python
df = pd.read_csv('./pokemon_data.csv')
df.head()
```

```python
df.groupby(['Type 1']).mean()
```

```python
df.groupby(['Type 1']).mean().sort_values('Attack', ascending=False)
```

```python
df.groupby(['Type 1']).sum()
```

```python
df.groupby(['Type 1']).count()
```

```python
df['count'] = 1
df.groupby(['Type 1', 'Type 2']).count()['count']
```

## Working with large amounts of data

```python
for df in pd.read_csv('modified.csv', chunksize=5):
    print("Chunk DF")
    print(df)
```

```python
new_df = pd.DataFrame(columns=df.columns)
for df in pd.read_csv('modified.csv', chunksize=5):
    results = df.groupby(['Type 1']).count()    # divide and conquer
    new_df = pd.concat([new_df, results])       # merge
```
